```python
import pandas as pd
import re
import datetime as dt
```

```python
pd.options.display.max_columns = 500
```

```python
def format_mediciones(medicion):
    if type(medicion) == str:
        return float(re.sub('\<|\>', '', medicion))
    else:
        return medicion
```

```python
def format_fecha(row):
    return dt.datetime.strptime(f'{row["Fecha de Muestreo"]} {row["Hora de Muestreo"]}', '%d-%m-%Y %H:%M Hs')
```

```python
def format_puntofeatures(feature):
    if type(feature) == str:
        if 'sin datos' in feature.lower():
            return None
        else:
            numbers = re.findall('[0-9\.]+', feature)
            if numbers:
                return float(numbers[0])
            else:
                return None
    else:
        return feature
```

```python
df = pd.read_csv('data/mediciones.csv')

for col in df.columns[14:]:
    df[col] = df[col].apply(format_mediciones)

df.insert(0, 'fecha', df[['Fecha de Muestreo', 'Hora de Muestreo']].apply(format_fecha, axis=1))

for col in ['Altura', 'Velocidad media', 'Caudal']:
    df[col] = df[col].apply(format_puntofeatures)
```

```python
selected_columns = ['id_syscali01', 'Responsable', 'Campaña','fecha', 'latitud', 'longitud', 'Río:', 'Altura', 'Velocidad media', 'Caudal'] + df.columns[14:].tolist()
name_columns = ['id', 'responsable', 'campaña', 'fecha', 'latitud', 'longitud', 'rio', 'altura (msnm)', 'velocidad_media (m/s)', 'caudal (m3/s)'] + df.columns[14:].str.lower().str.strip().tolist()
df = df[selected_columns]
df.columns = name_columns
```
